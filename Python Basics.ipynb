## **Theory Questions:-**


1. What is Python, and why is it popular?

- Python is a high-level, interpreted programming language known for its simplicity and readability. It is popular because of various reason:-
- It is versatile.
- It has lots of libraries.
- Easy to Learn and Read.
- Dynamic Typing.
- Community Support.


2. What is an interpreter in Python?

- An interpreter in Python is a program that reads and executes Python code line by line.


3. What are pre-defined keywords in Python?

- Predefined keywords in Python are special reserved words that have a fixed meaning and a specific purpose in the Python language.


4. Can keywords be used as variable names?

- No, keywords cannot be used as variable names in Python.
   -  Because keywords are reserved words that Python uses for its own syntax and rules.
  - If you try to use a keyword as a variable name, Python will give a SyntaxError.


5. What is mutability in Python?

- Mutability in Python means whether an object’s value can be changed after it is created.

6. Why are lists mutable, but tuples are immutable?

- Lists can change because they are designed with a flexible memory structure that allows adding, removing, or modifying items.
- Tuples cannot change because they use a fixed memory structure created for safety, speed, and storing constant data.


7. What is the difference between “==” and “is” operators in Python?

- "==" comparison operator to check if values are same or not.
- "is" whether both variables point to the same object in memory.


8. What are logical operators in Python?

- Logical operators in Python are used to combine multiple conditions and return either True or False.
Python has three logical operators:
    - and:- Returns True only if both conditions are True.
    - or:-  Returns True if at least one condition is True.
    - not:- changes true to false and vise-versa.


9. What is type casting in Python?

- Type casting means converting one data type into another, like int() → integer, float() → decimal, str() → string.


10. What is the difference between implicit and explicit type casting?

- An implicit type of conversion is a conversion performed by the python interpreter without programmer intervention.(Automatic Conversion).
- An Explicit type conversion is user - defined that forces an expression or to be of specific data type.


11. What is the purpose of conditional statements in Python?

- Conditional statements allow a program to make decisions and execute code only when a condition is true.
They help control the flow of the program based on different situations.


12. How does the elif statement work?

- elif checks another condition only if the previous (if) condition was false.
It helps you test multiple conditions one by one until one becomes true.


13. What is the difference between for and while loops?

- For loop is used when you already know how many times you want to repeat something.
- While loop is used when you don’t know the number of repeats and want to run the loop until a condition becomes false.


14. Describe a scenario where a while loop is more suitable than a for loop.

- while loop is more suitable when you don’t know how many times the loop should run and you want it to continue until a specific condition becomes false.
   - Example: keeping a login screen running until the user enters the correct password.

```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify(num, lst):
    num = num + 5        # Attempt to modify immutable object
    lst.append(50)       # Modify mutable object
    return num, lst

# Immutable argument (int)
number = 30

# Mutable argument (list)
my_list = [101, 102, 103]

print("Before function call:")
print("\nnumber:", number)
print("\nmy_list:", my_list)

result_num, result_list = modify(number, my_list)

print("\nAfter function call:")
print("\nnumber:", number)          # int argument unchanged(IMMUTABLE)
print("\nmy_list:", my_list)        # list argument changed(MUTABLE)
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

sum_of_evens = 0
for num in range(1, 51):
    if num % 2 == 0:
        sum_of_evens += num

print(f"The sum of all even numbers between 1 and 50 is: {sum_of_evens}")
```

```python
# 14. Write a program to reverse a string using a while loop.

s = input("Enter a string: ")
reversed_str = ""

i = len(s) - 1

while (i >= 0):
    reversed_str += s[i]
    i -= 1

print("Reversed string:", reversed_str)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
fact = 1
i = 1

while (i <= num):

    fact = fact * i
    i = i + 1

print("Factorial of", num, "is", fact)

```
